```python
def outlierremover(sourcedf):
    # Identifying the numeric cols
    col_list = list(sourcedf.select_dtypes(include=['int64', 'float64']).columns)

    for col in col_list:
        Q1 = sourcedf[col].quantile(0.25)
        Q3 = sourcedf[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        cleandf = sourcedf[(sourcedf[col] >= lower) & (sourcedf[col] <= upper)]
        outlierdf = sourcedf[(sourcedf[col] < lower) & (sourcedf[col] > upper)]
    return cleandf, outlierdf
```

```python
clean, source = outlierremover(df)
```

```python
def quicksort(inlist):
    listlen = len(inlist)
    while not all(inlist[index] < inlist[index + 1] for index in range(0, listlen-1)):
        for index in range(0, listlen-1):
            i = inlist[index]
            j = inlist[index + 1]
            if i > j:
                inlist[index], inlist[index + 1] = inlist[index + 1], inlist[index]
    return inlist
```
